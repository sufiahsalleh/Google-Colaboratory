Plot which country will make Malaysia as their top destination

```python
#import library that is needed
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#read CSV file
data = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019.csv')
data
```

```python
#Create TOTAL column tu calculate sum
#axis: If we want to aggregate the columns, then we’ll use axis=1. For rows we’ll use axis=0.
#numeric_only = take only numeric columns.

data['total'] = data.sum(axis=1, numeric_only= True)
data
```

```python
#read new CSV file with TOTAL column
newdata = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019_with_Total.csv')
newdata
```

```python
#Declare data that is needed to plot graph
total = newdata['total']
cName = newdata['Country Code']
```

```python
#plot graph
fig = plt.figure()
myaxes = fig.add_axes([10.0,4.0,40.0,8.0])
myaxes.plot(cName,total)
#view in new tab for clearer images
```
