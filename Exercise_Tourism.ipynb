Plot which country will make Malaysia as their top destination

```python
#import library that is needed
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#read CSV file
data = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019.csv')
data
```

```python
#Create TOTAL column tu calculate sum
#axis: If we want to aggregate the columns, then we’ll use axis=1. For rows we’ll use axis=0.
#numeric_only = take only numeric columns.

data['total'] = data.sum(axis=1, numeric_only= True)
data
```

```python
#read new CSV file with TOTAL column
newdata = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019_with_Total.csv')
newdata
```

```python
#Declare data that is needed to plot graph
x = list(newdata['Country Code'])
y = list(newdata['total'])
fig, ax = plt.subplots(figsize=(200, 20))
plt.xlabel('Country Code')
plt.ylabel('total')
ax.plot(x,y, linewidth = 2)
plt.show()
#open in new tab for clearer view
```

```python
#to view TOTAL column information
newdata['total'].describe()
```

```python
#to view which index have the highest TOTAL value
high_total = newdata['total'].idxmax()
high_total
# The country that will make Malaysia their top destination is world (WLD)
```
